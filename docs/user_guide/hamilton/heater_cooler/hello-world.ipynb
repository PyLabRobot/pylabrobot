# Hamilton Heater Cooler (HHC)

The Hamilton Heater Cooler is a Peltier-based [temperature controller](../../capabilities/temperature-control) for microplates. It connects to a STAR liquid handler via RS-232 through a TCC port.

| Spec | Value |
|---|---|
| PLR Name | `HamiltonHeaterCooler` |
| Temperature range | 0 to 110 °C |
| Cooling | Active (Peltier) |
| Communication | STAR TCC port (RS-232) |

See the [OEM page](https://www.hamiltoncompany.com/temperature-control/hamilton-heater-cooler) for hardware details.

## Setup

The HHC is accessed through a STAR backend. Pass the backend and the TCC device number (1-based) to create the device.

```python
from pylabrobot.hamilton.heater_cooler import HamiltonHeaterCooler

# Assumes `star` is an already-setup STAR backend
hhc = HamiltonHeaterCooler(name="hhc", star_backend=star, device_number=1)
await hhc.setup()
```

## Temperature control

The HHC exposes a {class}`~pylabrobot.capabilities.temperature_controlling.temperature_controller.TemperatureController` on `hhc.tc`. For the full API, see [Temperature Control](../../capabilities/temperature-control).

```python
await hhc.tc.set_temperature(37.0)
await hhc.tc.wait_for_temperature(tolerance=0.5)
```

```python
current = await hhc.tc.request_temperature()
print(f"{current:.1f} \u00b0C")
```

```python
await hhc.tc.deactivate()
```

## Teardown

```python
await hhc.stop()
```
