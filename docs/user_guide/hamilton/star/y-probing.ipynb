# Y-probing

With PyLabRobot, you can probe the y position of any object on a STAR(let) deck. See also [z-probing](./z-probing) for doing the same in the z direction.

```{warning}
This example uses the teaching tips. These are metal tips that are not forgiving. Be particularly careful when moving the channels around to avoid collisions.
```

```{note}
The y-probing methods (`prepare_for_manual_channel_operation`, `move_channel_x`, `move_channel_y`, `clld_probe_y_position_using_channel`) have **not** been migrated to the new API yet. They are only available on the legacy `STARBackend`. This notebook uses the new `STAR` device for setup, but the actual probing calls use the legacy backend directly.
```

## Setup

Import {class}`~pylabrobot.hamilton.liquid_handlers.star.star.STAR` and one of the Hamilton deck classes.

````python
from pylabrobot.hamilton.liquid_handlers.star import STAR
from pylabrobot.resources.hamilton import STARDeck

deck = STARDeck()
star = STAR(deck=deck)
await star.setup()
````

## Capacitive y-probing using cLLD

If you are mapping a capacitive surface, you can use the cLLD sensor to detect the surface.

```{warning}
This example uses the teaching tips. These are metal tips that are not forgiving. Be particularly careful when moving the channels around to avoid collisions.
```

```{note}
The probing methods below have not been ported to the new API yet. Until they are, you must use the legacy `LiquidHandler` / `STARBackend` interface instead of the `STAR` device shown above. The legacy setup looks like this:

    from pylabrobot.liquid_handling import LiquidHandler
    from pylabrobot.liquid_handling.backends.hamilton import STARBackend
    from pylabrobot.resources import STARDeck

    star = STARBackend()
    lh = LiquidHandler(backend=star, deck=STARDeck())
    await lh.setup()
```

Pick up a teaching tip:

````python
teaching_tip = lh.deck.get_resource("teaching_tip_rack")["A1"]
await lh.pick_up_tips(teaching_tip)
````

Prepare channel 0 for manual operation and move it to the desired x/y position:

````python
await star.prepare_for_manual_channel_operation(0)

# TODO: change this to a position that works for you
await star.move_channel_x(0, 500)
await star.move_channel_y(0, 300)
````

Use `STARBackend.clld_probe_y_position_using_channel` to probe the y-position of a single point at the current xz plane. This function will slowly move the channel until the liquid level sensor detects a change in capacitance. The y-coordinate of the tip is then returned.

````python
await star.clld_probe_y_position_using_channel(
  channel_idx=0,
  probing_direction="forward",
)
````

````python
await star.clld_probe_y_position_using_channel(
  channel_idx=0,
  probing_direction="backward",
)
````

Return the teaching tip when done:

````python
await lh.return_tips()
````
