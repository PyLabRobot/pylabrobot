# Hamilton HEPA Fan

The Hamilton HEPA Fan is an air filtration attachment for STAR and STARlet liquid handling workstations. It filters ambient air and supplies clean air to the deck interior, protecting samples from airborne contamination.

| Detail | Value |
|---|---|
| PLR Name | `HamiltonHepaFan` |
| Communication | Serial (FTDI) / USB-A |
| VID:PID | `0856:ac11` |
| Capabilities | [Fan control](../../capabilities/fan-control) |

**Physical setup:** Place the HEPA fan on top of the STAR(let) chassis. Connect the power cord (IEC C13) and a USB cable (USB-B at the fan, USB-A at the control PC). On older models, verify the voltage switch matches your mains voltage.

## Setup

```python
from pylabrobot.hamilton.only_fans import HamiltonHepaFan

hhfan = HamiltonHepaFan(name="hepa_fan")
await hhfan.setup()
```

If you have multiple fans, pass `device_id` to select a specific FTDI device:

```python
hhfan = HamiltonHepaFan(name="hepa_fan", device_id="ABC123")
await hhfan.setup()
```

## Fan control

The HEPA fan exposes a {class}`~pylabrobot.capabilities.fan_control.fan_control.Fan` capability on `hhfan.fan`. For the full API, see [Fan control](../../capabilities/fan-control).

Run at a given intensity (returns immediately):

```python
await hhfan.fan.turn_on(intensity=100)
```

```python
await hhfan.fan.turn_off()
```

Run for a fixed duration (blocks until done):

```python
await hhfan.fan.turn_on(intensity=80, duration=30)
```

## Teardown

```python
await hhfan.stop()
```
