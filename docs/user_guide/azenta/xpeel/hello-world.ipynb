# Azenta XPeel

The XPeel is an automated plate seal remover from Azenta. It supports:

- [Peeling](../../capabilities/peeling) (automated de-seal cycles with configurable speed and adhesion time)

The device communicates over RS-232 serial. Connect via a serial or USB-to-serial adapter.

## Setup

```python
from pylabrobot.azenta import XPeel

xpeel = XPeel(name="xpeel", port="/dev/ttyUSB0")  # replace with your port
await xpeel.setup()
```

## Peeling

The XPeel exposes a {class}`~pylabrobot.capabilities.peeling.peeling.Peeler` on `xpeel.peeler`. For the full API, see [Peeling](../../capabilities/peeling).

```python
await xpeel.peeler.peel()
```

The peel cycle can be customized with {class}`~pylabrobot.azenta.xpeel.XPeelPeelerBackend.PeelParams`:

- `begin_location`: starting position offset in mm (-2, 0, 2, or 4)
- `fast`: use faster peel speed
- `adhere_time`: tape adhesion time in seconds (2.5, 5.0, 7.5, or 10.0)

```python
from pylabrobot.azenta import XPeelPeelerBackend

await xpeel.peeler.peel(
    backend_params=XPeelPeelerBackend.PeelParams(
        begin_location=2,
        fast=True,
        adhere_time=5.0,
    )
)
```

## Driver utilities

The underlying {class}`~pylabrobot.azenta.xpeel.XPeelDriver` is available on `xpeel.driver` and exposes additional hardware commands.

Check for a seal on the plate:

```python
result = await xpeel.driver.seal_check()
print(result)  # "seal_detected", "no_seal", or "plate_not_detected"
```
