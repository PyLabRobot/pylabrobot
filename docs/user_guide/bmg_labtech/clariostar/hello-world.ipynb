# BMG Labtech CLARIOstar

The CLARIOstar is a multi-mode plate reader from BMG Labtech. It supports:

- [Absorbance](../../../capabilities/absorbance) (OD or transmittance)
- [Luminescence](../../../capabilities/luminescence)
- [Fluorescence](../../../capabilities/fluorescence) (not yet implemented in PLR)

Additional hardware features include a motorized loading tray, temperature control, and shaking.

| Model | PLR Name |
|---|---|
| CLARIOstar / CLARIOstar Plus | `CLARIOstar` |

- [OEM Link](https://www.bmglabtech.com/en/clariostar-plus/)
- **Communication**: USB (FTDI, VID 0x0403 / PID 0xBB68) at 125 000 baud
- **Communication level**: Firmware

## Physical setup

The CLARIOstar requires two cable connections:

1. Power cord (standard IEC C13)
2. USB cable (USB-B with security screws at the CLARIOstar end; USB-A at the control PC end)

If you have a plate stacking unit, an additional RS-232 port is available on the CLARIOstar.

## Setup

````python
from pylabrobot.bmg_labtech import CLARIOstar

reader = CLARIOstar(name="clariostar")
await reader.setup()
````

```{note}
On `setup()` the device performs its initialization routine. Pass a `device_id` to `CLARIOstar()` if you have multiple FTDI devices connected.
```

## Loading tray

Open and close the motorized plate tray:

````python
await reader.open()
````

````python
# load a plate (manually or via robotic arm)
await reader.close()
````

## Absorbance

The absorbance capability is exposed as `reader.absorbance`. For the full API, see [Absorbance](../../../capabilities/absorbance).

Use {class}`~pylabrobot.bmg_labtech.clariostar.absorbance_backend.CLARIOstarAbsorbanceParams` to select between OD and transmittance output.

````python
from pylabrobot.resources import Cor_96_wellplate_360ul_Fb

plate = Cor_96_wellplate_360ul_Fb(name="plate")
reader.plate_holder.assign_child_resource(plate)

results = await reader.absorbance.read_absorbance(plate, wavelength=450)
````

To get raw transmittance values instead of OD:

````python
from pylabrobot.bmg_labtech import CLARIOstarAbsorbanceParams

results = await reader.absorbance.read_absorbance(
    plate,
    wavelength=450,
    backend_params=CLARIOstarAbsorbanceParams(report="transmittance"),
)
````

## Luminescence

The luminescence capability is exposed as `reader.luminescence`. For the full API, see [Luminescence](../../../capabilities/luminescence).

````python
results = await reader.luminescence.read_luminescence(plate, focal_height=13.0)
````

## Fluorescence

The fluorescence capability is exposed as `reader.fluorescence`. For the full API, see [Fluorescence](../../../capabilities/fluorescence).

```{note}
Fluorescence reading is not yet implemented in the CLARIOstar driver.
```
