# Fluorescence

{class}`~pylabrobot.capabilities.plate_reading.fluorescence.fluorescence.Fluorescence` reads fluorescence intensity from microplates.

## When to use

Use this for fluorescence-based assays (GFP expression, DNA quantification with intercalating dyes, fluorescent ELISAs, etc.).

## Walkthrough

````python
from pylabrobot.capabilities.plate_reading.fluorescence import Fluorescence
from pylabrobot.capabilities.plate_reading.fluorescence.chatterbox import FluorescenceChatterboxBackend
from pylabrobot.resources import Cor_96_wellplate_360ul_Fb

fluorescence = Fluorescence(backend=FluorescenceChatterboxBackend())
await fluorescence._on_setup()

plate = Cor_96_wellplate_360ul_Fb(name="plate")
````

````python
results = await fluorescence.read(
    plate=plate,
    excitation_wavelength=485,
    emission_wavelength=528,
    focal_height=8.5,
)
print(f"A1 fluorescence: {results[0].data[0][0]} RFU")
````
