# Shaking

{class}`~pylabrobot.capabilities.shaking.shaking.Shaker` controls orbital and linear shakers, with optional plate locking.

## When to use

Use this for mixing samples, resuspending pellets, or keeping suspensions homogeneous during incubation.

## Walkthrough

````python
from pylabrobot.capabilities.shaking import Shaker
from pylabrobot.capabilities.shaking.chatterbox import ShakerChatterboxBackend

shaker = Shaker(backend=ShakerChatterboxBackend())
await shaker._on_setup()
````

Shake for a fixed duration (blocks for the full 60 seconds):

````python
await shaker.shake(speed=500, duration=2)  # 2 seconds for demo
````

Or start shaking indefinitely (returns immediately), then stop manually:

````python
await shaker.shake(speed=300)
# ... do other things ...
await shaker.stop_shaking()
````

Lock/unlock the plate (when the backend supports it):

````python
await shaker.lock_plate()
await shaker.shake(speed=1000, duration=1)
await shaker.unlock_plate()
````

## Tips and gotchas

- **With `duration`: blocks. Without: returns immediately.**
- **Auto-locking.** When the backend supports it, `shake()` automatically locks before shaking and unlocks after (when `duration` is set).
- **Speed is in RPM.**
