# Syringe dispensing

{class}`~pylabrobot.capabilities.bulk_dispensers.syringe.syringe.SyringeDispensing` controls syringe pump dispensing systems for precise, low-volume reagent delivery into microplates.

Syringe dispensers aspirate a fixed volume into a syringe barrel, then dispense it with high accuracy. They are ideal for expensive reagents, substrates, or detection reagents where precision matters.

## Walkthrough

```python
from pylabrobot.capabilities.bulk_dispensers.syringe import SyringeDispensing
from pylabrobot.capabilities.bulk_dispensers.syringe.chatterbox import (
    SyringeDispensingChatterboxBackend,
)
from pylabrobot.resources.corning.plates import Cor_96_wellplate_360ul_Fb

syringe = SyringeDispensing(backend=SyringeDispensingChatterboxBackend())
await syringe._on_setup()

plate = Cor_96_wellplate_360ul_Fb("demo_plate")
```

### Dispensing

Pass a `volumes` dict mapping **1-indexed column numbers** to volumes in uL.

```python
await syringe.dispense(
    plate=plate,
    volumes={col: 50.0 for col in range(1, 13)},
)
```

### Priming

Prime the syringe before dispensing.

```python
await syringe.prime(plate=plate, volume=5000.0)
```

## Peristaltic vs syringe

| | Peristaltic | Syringe |
|---|---|---|
| **Volume range** | Medium--high | Low--medium |
| **Accuracy** | Good | High |
| **Throughput** | High | Lower |
| **Purge needed** | Yes | No |
| **Typical use** | Media, wash buffer | Detection reagents, substrates |

Some devices (like the BioTek EL406) have both. Use the one that matches your volume and accuracy requirements.
