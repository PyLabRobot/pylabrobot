# Peristaltic dispensing

{class}`~pylabrobot.capabilities.bulk_dispensers.peristaltic.peristaltic.PeristalticDispensing` controls peristaltic pump dispensing systems for bulk reagent delivery into microplates.

Peristaltic dispensers use a rotating pump head to push fluid through tubing. They offer high throughput and are ideal for media, wash buffer, and large-volume reagent addition.

## Walkthrough

```python
from pylabrobot.capabilities.bulk_dispensers.peristaltic import PeristalticDispensing
from pylabrobot.capabilities.bulk_dispensers.peristaltic.chatterbox import (
    PeristalticDispensingChatterboxBackend,
)
from pylabrobot.resources.corning.plates import Cor_96_wellplate_360ul_Fb

peristaltic = PeristalticDispensing(backend=PeristalticDispensingChatterboxBackend())
await peristaltic._on_setup()

plate = Cor_96_wellplate_360ul_Fb("demo_plate")
```

### Dispensing

Pass a `volumes` dict mapping **1-indexed column numbers** to volumes in uL.

```python
# Uniform volume across all 12 columns
await peristaltic.dispense(
    plate=plate,
    volumes={col: 50.0 for col in range(1, 13)},
)
```

```python
# Different volumes per column
await peristaltic.dispense(
    plate=plate,
    volumes={1: 10.0, 2: 20.0, 3: 30.0},
)
```

### Priming

Prime the fluid lines before dispensing to fill tubing with reagent. Specify either `volume` (uL) or `duration` (seconds), depending on what the hardware supports.

```python
await peristaltic.prime(plate=plate, volume=500.0)
```

### Purging

Purge (empty) the fluid lines after dispensing to clear the tubing.

```python
await peristaltic.purge(plate=plate, volume=500.0)
```

## Tips and gotchas

- **Always prime before dispensing.** Air in the tubing causes inaccurate volumes.
- **Purge after dispensing** to prevent reagent from drying in the lines.
- **Columns are 1-indexed.** `{1: 50.0}` sets column 1, not column 0.
- **Only columns in the dict are set.** Columns not in `volumes` retain their previous setting on the instrument.
