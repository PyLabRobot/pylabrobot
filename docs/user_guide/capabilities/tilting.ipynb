# Tilting

{class}`~pylabrobot.capabilities.tilting.tilting.Tilter` controls tilt modules for angling plates or other labware.

## When to use

Use this to tilt plates for aspiration from low-volume wells, gravity-based liquid removal, or bead settling.

## Walkthrough

````python
from pylabrobot.capabilities.tilting import Tilter
from pylabrobot.capabilities.tilting.chatterbox import TilterChatterboxBackend

tilter = Tilter(backend=TilterChatterboxBackend())
await tilter._on_setup()
````

````python
# Tilt to an absolute angle (0 = horizontal)
await tilter.set_angle(15)
print(f"angle: {tilter.absolute_angle}")
````

````python
# Tilt by a relative amount
await tilter.tilt(-5)
print(f"angle: {tilter.absolute_angle}")
````

````python
# Return to horizontal
await tilter.set_angle(0)
print(f"angle: {tilter.absolute_angle}")
````
