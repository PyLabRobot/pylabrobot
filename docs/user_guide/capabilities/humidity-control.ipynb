# Humidity Control

{class}`~pylabrobot.capabilities.humidity_controlling.humidity_controller.HumidityController` reads and sets relative humidity for incubators and environmental chambers.

## Walkthrough

````python
from pylabrobot.capabilities.humidity_controlling import HumidityController
from pylabrobot.capabilities.humidity_controlling.chatterbox import HumidityControllerChatterboxBackend

humid = HumidityController(backend=HumidityControllerChatterboxBackend())
await humid._on_setup()
````

````python
await humid.set_humidity(0.95)  # 95% RH
current = await humid.request_humidity()
print(f"Humidity: {current * 100:.1f}%")
````

## Tips and gotchas

- **Humidity is a fraction (0.0--1.0)**, not a percentage.
- **Some backends are read-only.** If `supports_humidity_control` is `False`, calling `set_humidity` raises `ValueError`.
