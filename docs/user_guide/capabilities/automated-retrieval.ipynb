# Automated Retrieval

{class}`~pylabrobot.capabilities.automated_retrieval.automated_retrieval.AutomatedRetrieval` controls automated storage systems (carousels, plate hotels, incubators with internal storage) that can fetch and store plates.

## When to use

Use this to programmatically retrieve plates from storage for processing and return them when done.

## Walkthrough

````python
from pylabrobot.capabilities.automated_retrieval import AutomatedRetrieval
from pylabrobot.capabilities.automated_retrieval.chatterbox import AutomatedRetrievalChatterboxBackend
from pylabrobot.resources import Cor_96_wellplate_360ul_Fb

retrieval = AutomatedRetrieval(backend=AutomatedRetrievalChatterboxBackend())
await retrieval._on_setup()

plate = Cor_96_wellplate_360ul_Fb(name="my_plate")
````

````python
await retrieval.fetch_plate_to_loading_tray(plate)
````
