# Automated Retrieval

{class}`~pylabrobot.capabilities.automated_retrieval.automated_retrieval.AutomatedRetrieval` controls automated storage systems (carousels, plate hotels, incubators with internal storage) that can fetch and store plates.

## When to use

Use this to programmatically retrieve plates from storage for processing and return them when done.

## Walkthrough

````python
from pylabrobot.capabilities.automated_retrieval import AutomatedRetrieval
from pylabrobot.capabilities.automated_retrieval.chatterbox import AutomatedRetrievalChatterboxBackend
from pylabrobot.resources import Coordinate, Cor_96_wellplate_360ul_Fb, PlateCarrier, PlateHolder

# The capability owns the storage racks and the loading tray, and implements the
# site bookkeeping (free-site counting, lookup, selection). A device that composes
# AutomatedRetrieval passes these in; here we build them by hand.
site = PlateHolder(name="site_0", size_x=127.76, size_y=85.48, size_z=30, pedestal_size_z=0)
site.location = Coordinate(0, 0, 0)
rack = PlateCarrier(name="rack_0", size_x=135, size_y=95, size_z=400, sites={0: site})
loading_tray = PlateHolder(
    name="loading_tray", size_x=127.76, size_y=85.48, size_z=0, pedestal_size_z=0
)

retrieval = AutomatedRetrieval(
    backend=AutomatedRetrievalChatterboxBackend(),
    racks=[rack],
    loading_tray=loading_tray,
)
await retrieval._on_setup()

# place a plate into storage so we have something to retrieve
site.assign_child_resource(Cor_96_wellplate_360ul_Fb(name="my_plate"))
print("free sites:", retrieval.get_num_free_sites())
````

````python
# fetch a stored plate onto the loading tray by name, then store it back into a free site
plate = await retrieval.fetch_plate_to_loading_tray("my_plate")
await retrieval.take_in_plate("smallest")
print(retrieval.summary())
````
