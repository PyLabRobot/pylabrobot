# Centrifuging

{class}`~pylabrobot.capabilities.centrifuging.centrifuging.Centrifuge` controls automated centrifuges with door, bucket, and spin management.

## When to use

Use this for pelleting cells, separating phases, or any protocol step that requires centrifugal force.

## Walkthrough

````python
from pylabrobot.capabilities.centrifuging import Centrifuge
from pylabrobot.capabilities.centrifuging.chatterbox import CentrifugeChatterboxBackend
from pylabrobot.resources import ResourceHolder

bucket1 = ResourceHolder(name="bucket1", size_x=100, size_y=100, size_z=50)
bucket2 = ResourceHolder(name="bucket2", size_x=100, size_y=100, size_z=50)

centrifuge = Centrifuge(
    backend=CentrifugeChatterboxBackend(),
    buckets=(bucket1, bucket2),
)
await centrifuge._on_setup()
````

````python
await centrifuge.open_door()
# ... load plate into bucket via arm ...
await centrifuge.close_door()
await centrifuge.lock_door()
````

````python
# Spin at 500 x g for 2 seconds (blocks until complete)
await centrifuge.spin(g=500, duration=2)
````

````python
await centrifuge.unlock_door()
await centrifuge.open_door()

# Bucket position is unknown after spinning
print(f"at_bucket: {centrifuge.at_bucket}")

# Rotate to a known position before unloading
await centrifuge.go_to_bucket1()
print(f"at_bucket: {centrifuge.at_bucket}")
````

## Tips and gotchas

- **`spin()` blocks** for the entire cycle (ramp up + time at speed + ramp down).
- **`duration` is time at speed**, excluding ramp-up and ramp-down.
- **Bucket position is unknown after spinning.** `at_bucket` resets to `None` after `spin()`.
