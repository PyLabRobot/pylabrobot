# Luminescence

{class}`~pylabrobot.capabilities.plate_reading.luminescence.luminescence.Luminescence` reads luminescence from microplates.

## When to use

Use this for luminescence-based assays (luciferase reporters, ATP quantification, chemiluminescent ELISAs, etc.).

## Walkthrough

````python
from pylabrobot.capabilities.plate_reading.luminescence import Luminescence
from pylabrobot.capabilities.plate_reading.luminescence.chatterbox import LuminescenceChatterboxBackend
from pylabrobot.resources import Cor_96_wellplate_360ul_Fb

luminescence = Luminescence(backend=LuminescenceChatterboxBackend())
await luminescence._on_setup()

plate = Cor_96_wellplate_360ul_Fb(name="plate")
````

````python
results = await luminescence.read(plate=plate, focal_height=8.5)
print(f"A1 luminescence: {results[0].data[0][0]} RLU")
````
