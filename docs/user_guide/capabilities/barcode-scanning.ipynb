# Barcode Scanning

{class}`~pylabrobot.capabilities.barcode_scanning.barcode_scanning.BarcodeScanner` reads barcodes from labware.

## When to use

Use this to identify plates, tubes, or other labware by their barcode during automated workflows -- for tracking, verification, or LIMS integration.

## Walkthrough

````python
from pylabrobot.capabilities.barcode_scanning import BarcodeScanner
from pylabrobot.capabilities.barcode_scanning.chatterbox import BarcodeScannerChatterboxBackend

scanner = BarcodeScanner(backend=BarcodeScannerChatterboxBackend())
await scanner._on_setup()
````

````python
barcode = await scanner.scan()
print(f"Scanned: {barcode}")
````

## Serial scanners

Scanners configured as RS-232 or USB virtual COM devices can use {class}`~pylabrobot.capabilities.barcode_scanning.serial.SerialBarcodeScanner`. Configure the scanner to emit a line ending such as carriage return or newline after each barcode. If the scanner documents a serial trigger command, pass it as `trigger_command`; otherwise `scan()` passively waits for the next barcode line.

````python
from pylabrobot.capabilities.barcode_scanning import SerialBarcodeScanner

serial_scanner = SerialBarcodeScanner(port="COM5", baudrate=115200)
await serial_scanner.setup()
barcode = await serial_scanner.barcode_scanning.scan(read_time=5)
await serial_scanner.stop()
````
