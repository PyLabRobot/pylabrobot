# Pumping

{class}`~pylabrobot.capabilities.pumping.pumping.Pump` controls peristaltic and syringe pumps for fluid delivery.

## When to use

Use this for bulk liquid delivery, reagent dispensing, or continuous flow applications where pipetting is impractical.

## Walkthrough

````python
from pylabrobot.capabilities.pumping import Pump
from pylabrobot.capabilities.pumping.chatterbox import PumpChatterboxBackend

pump = Pump(backend=PumpChatterboxBackend())
await pump._on_setup()
````

````python
# Run for a fixed number of revolutions (blocks)
await pump.run_revolutions(10)
````

````python
# Run continuously (returns immediately), then stop
await pump.run_continuously(50)
await pump.halt()
````

````python
# Run for a fixed duration (blocks)
await pump.run_for_duration(speed=30, duration=2)
````

## Calibration

To use `pump_volume`, the capability needs a {class}`~pylabrobot.capabilities.pumping.calibration.PumpCalibration`. The calibration converts a (speed, volume) pair into either a duration or a number of revolutions.

## Tips and gotchas

- **`run_continuously` returns immediately.** The pump runs until you call `halt()`.
- **`run_revolutions` and `run_for_duration` block.**
- **`pump_volume` requires a calibration.** Raises `NotCalibratedError` if none is set.
