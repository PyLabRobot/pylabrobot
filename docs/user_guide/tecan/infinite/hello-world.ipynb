# Tecan Infinite 200 PRO

The Tecan Infinite 200 PRO is a multimode microplate reader that supports:

- [Absorbance](../../capabilities/absorbance) (230--1000 nm)
- [Fluorescence](../../capabilities/fluorescence) (230--850 nm excitation/emission)
- [Luminescence](../../capabilities/luminescence)

This backend targets the Infinite "M" series (e.g., Infinite 200 PRO M Plex). The "F" series uses a different optical path and is not covered here.

````python
from pylabrobot.tecan.infinite import TecanInfinite200Pro

reader = TecanInfinite200Pro(name="reader")
await reader.setup()
````

````python
await reader.loading_tray.open()
````

Before closing, assign a plate to the reader. This determines the well positions for measurements.

````python
from pylabrobot.resources import Cor_96_wellplate_360ul_Fb
plate = Cor_96_wellplate_360ul_Fb(name="plate")
reader.plate_holder.assign_child_resource(plate)
````

````python
await reader.loading_tray.close()
````

## Absorbance

Read absorbance at a specified wavelength (230--1000 nm). For the full API, see [Absorbance](../../capabilities/absorbance).

````python
results = await reader.absorbance.read(plate=plate, wavelength=450)
results[0].data  # 2D array indexed [row][col]
````

### Backend-specific parameters

Use {class}`~pylabrobot.tecan.infinite.TecanInfiniteAbsorbanceParams` to configure flashes and bandwidth.

````python
from pylabrobot.tecan.infinite import TecanInfiniteAbsorbanceParams

results = await reader.absorbance.read(
    plate=plate,
    wavelength=450,
    backend_params=TecanInfiniteAbsorbanceParams(flashes=50, bandwidth=9.0),
)
````

## Fluorescence

Read fluorescence with specified excitation and emission wavelengths (230--850 nm). The focal height is in millimeters. For the full API, see [Fluorescence](../../capabilities/fluorescence).

````python
results = await reader.fluorescence.read(
    plate=plate,
    excitation_wavelength=485,
    emission_wavelength=528,
    focal_height=20.0,
)
results[0].data
````

## Luminescence

Read luminescence. The focal height is in millimeters. For the full API, see [Luminescence](../../capabilities/luminescence).

````python
results = await reader.luminescence.read(plate=plate, focal_height=20.0)
results[0].data
````

## Reading specific wells

You can specify a subset of wells to read instead of the entire plate.

````python
wells = plate.get_items(["A1", "A2", "B1", "B2"])
results = await reader.absorbance.read(plate=plate, wavelength=450, wells=wells)
````

## Teardown

````python
await reader.stop()
````

## Installation

The Infinite 200 PRO connects via USB. PyLabRobot uses `pyusb` for communication, which requires `libusb` on your system.

### macOS

```bash
brew install libusb
```

### Linux (Debian/Ubuntu)

```bash
sudo apt-get install libusb-1.0-0-dev
```

### Windows

Install [Zadig](https://zadig.akeo.ie/) and replace the Infinite's default USB driver with `WinUSB` or `libusb-win32`.
